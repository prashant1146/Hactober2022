## read_csv()
The read_csv() function facilitates the reading of a comma-separated values (csv) file into a Pandas DataFrame. All you have to do is specify the path to the file that you want it to read. It can also read files with delimiters other than commas, such as | or tab.

The data from the data source has been read into the Pandas DataFrame. You must update the path of the file you wish to read.

```python
import pandas as pd
```

```python
data=pd.read_csv('blog_dataset.csv')
```

```python
data
```

## head()

head(n) returns the top n rows of a dataset. df.head() returns the first 5 rows of the DataFrame by default. You can specify n as an integer if you want more or fewer rows.


```python
data.head(6)
```

## Tail()
tail(), like head(), retrieves the lowest n rows of a dataset. Head() and tail() allow you to quickly scan your dataset and determine whether data has been properly read into the DataFrame.

```python
data.tail()
```

## describe()
describe() produces descriptive statistics for data in a Pandas DataFrame or Series. It summarises the dataset's central tendency and dispersion. description() provides a short overview of the dataset.

describe() returns a list of various descriptive statistical metrics for each number column in our dataset. By setting the include property to 'all,' we can make the description include all columns, including those holding categorical data.

```python
data.describe()
```

## memory_usage()
memory_usage() returns a Pandas Series containing the memory usage (in bytes) of each column in a Pandas DataFrame. We can find out how much space each column takes by setting the deep property to True.

In a Pandas Series, the memory use of each column has been output. It is critical to understand a DataFrame's memory usage so that you can handle Python issues such as MemoryError.

```python
data.memory_usage(deep=True)
```

## astype()
To convert a Python object to a specific data type, use the astype() function. If your data is not kept in the proper format, this feature may be quite useful (data type). For instance, if Python inadvertently misunderstood your floating point values as strings, you can use the astype function to change them back (). You can also use astype to change an object datatype into a category ().

```python
data['Gender'] = data.Gender.astype('category')
```

## loc[:]
According to our needs, loc[:] enables access to a collection of rows and columns in a dataset, or a portion of the dataset. For instance, we can use loc[:] to access a dataset's last 2 rows and first 3 columns if that's all we need. Instead of using row and column numbers, we may also retrieve rows and columns based on labels.

```python
data.loc[0:4, ['Name', 'Age', 'State']]
```

## to_datetime()
to_datetime() changes a Python object's format to datetime. It will accept as arguments an integer, a floating point number, a list, a Pandas Series, or a Pandas DataFrame. When the dataset contains dates or time series values, to_datetime() is quite useful.

```python
data['DOB'] = pd.to_datetime(data['DOB'])
```

```python
data
```

## value_counts()
Using value_counts(), one can obtain a Pandas Series with the counts of distinct values. Think about a dataset that includes data on 5,000 of a company's clients. The value_counts() function will assist us in counting the instances of each distinct value within a Series. It can be used with columns that hold data like the state, the industry in which one works, or the age of the clientele.

```python
data['State'].value_counts()
```

```python
data['State'].value_counts(normalize=True).plot(kind='bar', title='State')
```

## drop_duplicates()
A Pandas DataFrame with duplicate rows eliminated is returned by drop_duplicates(). There is an option to maintain the first instance (record) of a duplicate or the final instance even when there are duplicates. The inplace and ignore index attributes are also options.

The use of inplace=True ensures that the modifications are made to the original dataset. By comparing the original dataset's and the changed dataset's shapes, you can confirm the changes (after dropping duplicates). You'll see that there are now 8 instead of 9 rows (because 1 duplicate has been dropped).

```python
data.drop_duplicates(inplace=True)
```

```python
data
```

## groupby()
A Pandas DataFrame can be grouped by one or more columns and subjected to a mathematical operation using the groupby() function. It is possible to easily summarise data with groupby().

```python
data.groupby(by='State').Salary.mean()
```

```python
data
```

## sort_values()
A Pandas DataFrame (or Pandas Series) column can be sorted by values in either ascending or descending order using the sort values() function. You can directly alter the original DataFrame by setting the inplace property to True.

```python
data.sort_values(by='Name', inplace=True)
```

```python
data
```

## fillna()
A large dataset will typically contain multiple entries that Python has labelled as NaN. NaN, which stands for "not a number," refers to records that the original data source did not populate. Pandas ensures that each entry can be distinguished from the others when populating the DataFrame with values.

By imputing these missing values with more suitable values, the fillna() function aids in replacing all NaN entries in a DataFrame or Series.

```python
data['City temperature'].fillna(38.5, inplace=True)
```

```python
data
```
