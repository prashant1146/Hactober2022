```python
import matplotlib.pyplot as pl
x=np.arange(6)
y=[30,40,50,60,70,80]
b=[25,45,60,75,55,93]
pl.title("BAR GRAPH")
pl.xlabel("X-axis")
pl.ylabel("Y-axis")
#legend is introduced
pl.bar(x,y,color=['yellow','yellow','yellow'],width=0.25,label="1st SEM")
pl.bar(x+0.25,b,color=['red','red','red'],width=0.25,label="2nd SEM")
pl.legend(loc="upper left")
pl.show()
```

```python
import matplotlib.pyplot as pl
x=np.arange(6)
y=[30,40,50,60,70,80]
b=[25,45,60,75,55,93]
pl.title("BAR GRAPH")
pl.xlabel("X-axis")
pl.ylabel("Y-axis")
#tick is introduced
pl.xticks(x,["ENG","HIN","CS","MATH","PHY","CHEM"])
pl.bar(x,y,color=['yellow','yellow','yellow'],width=0.25,label="1st SEM")
pl.bar(x+0.25,b,color=['red','red','red'],width=0.25,label="2nd SEM")
pl.legend(loc="upper left")  #location can me changed "upper left,lower left/right"
pl.show()
```
