## Important types of plots in Matplotlib
### Sub Plots-
The Matplotlib function subplots() is used to combine numerous plots into a single figure. It accepts different parameters, including the quantity of rows, columns, or sharex, sharey axes.

```python
# First create a grid of plots
fig, ax = plt.subplots(2,2,figsize=(15,10)) #this will create the subplots with 2 rows and 2 columns 
#and the second argument is size of the plot 
# Lets plot all the figures 
x1=[1960,1970,1980,1990,2000,2010]
ax[0][0].plot(x1, np.sin(x1), 'g') #row=0,col=0 
ax[0][1].plot(x1, np.cos(x1), 'y') #row=0,col=1
 
ax[1][0].plot(x1, np.sin(x1), 'b') #row=1,col=0
ax[1][1].plot(x1, np.cos(x1), 'red') #row=1,col=1
plt.tight_layout() 
#show the plots
plt.show()
```

### Line Plots-
To visualise the relationship between the x and y axes, use a line plot.

The x and y coordinates are plotted in a 2D hexagonal pattern using the plot() function in the Pyplot module of the Matplotlib toolkit. Several parameters can be sent to plot(), including plot(x, y, scalex, scaley, data, **kwargs).

The horizontal and vertical axes' coordinates are x and y, respectively; x values are optional and have the default value of range(len(y)).

The x-axis or y-axis can be automatically scaled using the scalex, scaley parameters, and their default value is true.

```python
#this line will create array of numbers between 1 to 10 of length 100 
#np.linspace(satrt,stop,num) 
x1 = np.linspace(0, 10, 50)  #line plot 
plt.plot(x1, np.sin(x1), '-',color='orange') 
plt.plot(x1, np.cos(x1), '--',color='b')
#give the name of the x and y  axis 
plt.xlabel('x label')
```

### Histogram-
A histogram is the most popular type of graph used to show frequency distributions. The first step in creating a histogram is to establish a bin of ranges. Next, divide the entire range of values into a series of intervals, and then count the values that fall within each interval. For plotting histograms, we can use the plt.hist() method, which accepts a variety of inputs, including data, bins, colour, etc.

x: The array's x-coordinate or sequence

bins: An integer indicating the desired number of bins for the graph.

range: the lower and higher bin ranges

density: a boolean-valued optional parameter.

Histtype is an optional parameter that can be used to construct a variety of histograms, including bars, bars stacked, steps, and filled steps.

```python
#draw random samples from random distributions.
x = np.random.normal(200, 10, 300)
#plot histograms
plt.hist(x) 
plt.show()
```

### Barplots

Barplots are mostly used to illustrate the relationship between categorical and numerical values. One axis in a bar chart represents a specific category of columns, and another axis shows the values or count of that category. Using the following line of code, barcharts can be plotted both vertically and horizontally:

x: used to represent the x-axis coordinates.

height: The bars' height.

width: The bars' overall width. 0.8 is the default value.

bottom: It's not required. It has the default value of None and is a y-coordinate for the bar.

centres and edges. its standard setting is centre

```python
#define array  
data= [5. , 25. , 50. , 20.,15.] 
plt.bar(range(len(data)), data,color='r') 
plt.show()
```

### Scatter Plot-
Dots are used for charting in scatter plots, which are intended to display relationships between variables. They can also be used to display relationships between two numerical variables.

Plotting is done using the Matplotlib library's scatter() technique.


```python
#create the x and y axis coordinates
x = np.array([5,7,8,11,12,9,6,7,2,17,2,9,4]) 
y = np.array([99,111,86,103,87,94,86,87,88,78,77,85,86]) 
plt.scatter(x, y)
plt.legend() 
plt.show()
```

### Pie Chart-
The percentage of the entire is displayed using a pie chart (or circular chart). Thus, it is employed whenever we wish to contrast several categories with one another. Pie() will accept a variety of parameters, including:

x: A list of arrays

labels: A list of strings that represent the names of the pie-chart slices.

The wedges are labelled with numerical values using autopct. The wedges will then contain the labels. Format is %1.2f%.

```python
#define the figure size
plt.figure(figsize=(8,8))
x = [30,45,25,10] 
#labels of the pie chart
labels = ['A','B','C','D'] 
plt.pie(x, labels=labels)
plt.show()
```

### Box Plot-
A box plot is used to display the dataset's overall overview or all of its numerical values. The minimum, first quartile, median, third quartile, and maximum are all included in the summary. Additionally, between the first and third quartiles, the median is present. Here, the y coordinates display the frequency distribution while the x axis displays the data values.

The following are the parameters used in box plots:

NumPy array of data

For the vertical and horizontal plots, vert will accept boolean values of true or false; the default value is True.

box widths: This will accept an array and sets of box widths.



labels: An array of strings used to set the dataset's labels

```python
#create the random values by using numpy 
values= np.random.normal(100, 120, 300) 
#creating the plot by boxplot() function which is avilable in matplotlib
plt.boxplot(values,patch_artist=True,vert=True) 
plt.show()
```

### Area Chart-
The area chart or area plot, which is based on the line plot, is used to graphically represent quantitative data. Plotting the area chart involves using the fill between() method.

Parameter:

The plot's x and y coordinates are represented by x,y. This requires an array with length n.

Interpolate is an optional boolean value. If so, extrapolate between the two lines to get the exact intersection's location.

Alpha, colour, facecolor, edgecolor, and linewidth are the **kwargs.

```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
y = [2, 38, 25, 18, 7, 14, 17, 20, 27, 30, 6, 1]
#plot the line of the given data
plt.plot(np.arange(12),y, color="blue", alpha=0.6, linewidth=2) 
#decorate thw plot by giving the labels 
plt.xlabel('Month', size=12) 
plt.ylabel('Turnover(Cr.)', size=12) #set y axis start with zero 
plt.ylim(bottom=0)
plt.show()
```

### Word Cloud-
The visual representation of the text data is a wordcloud. Typically, each word is a single unit, and the font size or colour indicates how important it is. In Python, word clouds are made using the wordcloud() method.

The wordcloud() will accept a number of arguments, including:

width: choose the canvas's width.

standard 400

height: choose the canvas's desired height.

standard 400

max words: the maximum amount of words; 200 is the default.

background color: The colour of the word-cloud image's background.

Black is the default colour.

After creating the word cloud object, you can call the generate function and pass the text data to create the word cloud.

```python
#import the libraries
from wordcloud import WordCloud 
import matplotlib.pyplot as plt 
from PIL import Image 
import numpy as np
#set the figure size . 
plt.figure(figsize=(10,15)) 
#dummy text.
text = '''Lorem Ipsum is simply dummy text of the printing and typesetting industry. 
        Lorem Ipsum has been the industry's standard dummy text ever since the 1500s, 
        when an unknown printer took a galley of type and scrambled it to make a type specimen book. 
        It has survived not only five centuries, but also the leap into electronic typesetting, 
        remaining essentially unchanged. It was popularised in the 1960s with the release of 
        Letraset sheets containing Lorem Ipsum passages, and more recently with desktop publishing 
        software like Aldus PageMaker including versions of Lorem Ipsum.'''


wordcloud = WordCloud( margin=0,colormap='BuPu').generate(text) 
#imshow() function in pyplot module of matplotlib library is used to display data as an image.
plt.imshow(wordcloud, interpolation='bilinear') 
plt.axis('off') 
plt.margins(x=0, y=0) 
plt.show()
```

### 3-D Graphs-
It's time to check out some sophisticated graphs, namely 3-D graphs, now that you've seen some simple ones. Matplotlib was initially designed for 2-dimensional graphs, however 3-D graphs were later added. Let's see how to use Matplotlib to plot a 3-D graph.

```python
from mpl_toolkits import mplot3d
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
fig = plt.figure()
ax = plt.axes(projection='3d')
```

Every single plot that we have seen using Matplotlib's 2-D plotting functionality may also be created as 3-D graphs. Check out a line plot in a 3-D plane as an example.

```python
ax = plt.axes(projection='3d')

# Data for a three-dimensional line
zline = np.linspace(0, 20, 1000)
xline = np.sin(zline)
yline = np.cos(zline)
ax.plot3D(xline, yline, zline, 'gray')

# Data for three-dimensional scattered points
zdata = 20 * np.random.random(100)
xdata = np.sin(zdata) + 0.1 * np.random.randn(100)
ydata = np.cos(zdata) + 0.1 * np.random.randn(100)
ax.scatter3D(xdata, ydata, zdata, c=zdata, cmap='Greens');
```
